 **Task** - 
Will be reading the data set and cropping the face and saving it inside the cropped folder 

```python
import os 
import cv2 
from keras.preprocessing import image 
import numpy as np 
import csv
 

# Haar Cascade face detection model
face_cascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))
```

I . **Tag** - 

    0 - with mask 
    1 - without mask 
    
II. **Task** - 
  
   Here , we will be reading the images from the google drive and the create a csv file to store . 
   
   The csv file have 3 columns - **usage , pixeles and tag**






```python

# from where all the images will be read 
# directory = 'drive/MyDrive/FaceMaskDetection/DateSet_Training' 
directory = 'drive/MyDrive/FaceMaskDetection/TestData1' 

columns = [ 'pixels' , 'tag' ]                                                    # Columns of the csv file 

data = []                                                                         # data in the csv file  

# traversing the dirs inside the path 
for dir in os.listdir(directory) : 
  
  # generating the name of the sub-directory 
  subdir = os.path.join( directory, dir )                                         # selecting a certain sub-directory 
  print( "Processing the direcotry : " , subdir , "\n\n") 

  # processing the files inside the directory 
  for files in os.listdir(subdir) :
    try :
      name = os.path.join( subdir , files )                                       # name of the file with full path 
      print(name)

      # read the image from the drive location (name) that is passed 
      img = cv2.imread( name , cv2.IMREAD_COLOR )

      # convertion to grayscale 
      gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

      # get face bounding box coordinates using Haar Cascade
      faces = face_cascade.detectMultiScale(gray_img)
      
      # Traversing over all the faces that are detected 
      for (x,y,w,h) in faces: 
        
        temp = [] 

        cropped_gray = gray_img[y:y+w,x:x+h]                                      # cropping the face from the image 
        
        cropped_gray = cv2.resize(cropped_gray,(48,48))                           # resizing the image 
        
        img_pixels = image.img_to_array(cropped_gray)                             # getting the image pixeles into array 
        
        img_pixels = list( img_pixels.flatten() )                                 # flattening the image in row major fashion 
        
        tag = 1 if 'without' in name else 0 

        data.append([ ''.join( str(i)+' ' for i in img_pixels ) , tag ])

    except Exception as err:
      # Errors will be thrown if the user does not have a webcam or if they do not
      # grant the page permission to access it.
      print(str(err))

print( len(data) )




```

Writing the data into CSV file and storing it as DataSet.csv

```python
# writing all the data in CSV file 

with open('drive/MyDrive/FaceMaskDetection/DataSet.csv' , 'w') as fp :

  csvWrite = csv.writer(fp)
  csvWrite.writerow( columns )          # writing the column names
  csvWrite.writerows( data )            # writing the data 
```
